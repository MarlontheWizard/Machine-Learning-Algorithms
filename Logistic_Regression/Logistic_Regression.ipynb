```python
#Author: Marlon Dominguez
#Machine Learning, Rutgers University
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

#load data
data = pd.read_csv('Salary_Data.csv')

#get features and store in seperate arrays
X = data[['YearsExperience']].values  # Features: Years of Experience (as 2D array)
y = data['Salary'].values             # Target: Salary

#fit using logistic regression from sklearn
model = LinearRegression()
model.fit(X, y)

#generate the predictions based on x 
y_pred = model.predict(X)

#Plot the results 
plt.scatter(X, y, color='blue', label='Data Points')  # Plot the actual data
plt.plot(X, y_pred, color='red', label='Regression Line')  # Plot the regression line
plt.title('Linear Regression: Salary vs. Years of Experience')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend()
plt.show()

#Print the values for salary 
print(f'Linear Regression Equation: Salary = {model.coef_[0]:.2f} * YearsExperience + {model.intercept_:.2f}')
```
